```python
import pandas as pd
import numpy as np
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import accuracy_score,confusion_matrix
data=pd.read_csv('/content/income(1) (1).csv',na_values=[" ?"])
data
```

```python
data.isnull().sum()
```

```python
missing=data[data.isnull().any(axis=1)]
missing
```

```python
data2 = data.dropna(axis=0)
data2
```

```python
sal=data['SalStat']
data2['SalStat']=data2['SalStat'].map({' less than or equal to 50,000':0,' greater than 50,000':1})
print(data2['SalStat'])
```

```python
sal2=data2['SalStat']
dfs=pd.concat([sal,sal2],axis=1)
dfs
```

```python
data2
```

```python
new_data=pd.get_dummies(data2, drop_first=True)
new_data
```

```python
columns_list=list(new_data.columns)

features=list(set(columns_list)-set(['SalStat']))
print(features)

y=new_data['SalStat'].values
print(y)
```

```python
x = new_data[features].values
print(x)
```

```python
train_x,test_x,train_y,test_y = train_test_split(x,y,test_size=0.3, random_state=0)
KNN_classifier=KNeighborsClassifier(n_neighbors = 5)
KNN_classifier.fit(train_x,train_y)
```

```python
prediction = KNN_classifier.predict(test_x)
confusionMmatrix = confusion_matrix(test_y, prediction)
print(confusionMmatrix)
```

```python
from sklearn.metrics import accuracy_score
acc_score = accuracy_score(test_y, prediction)
print(acc_score)
```

```python
print('Misclassified samples: %d' % (test_y != prediction).sum())
```

```python
import pandas as pd
import numpy as np
from scipy.stats import chi2_contingency
import seaborn as sns
tips=sns.load_dataset('tips')
tips.head()
```

```python
contingency_table=pd.crosstab(tips['sex'],tips['time'])
print(contingency_table)
```

```python
chi2, p, _, _ = chi2_contingency(contingency_table)
print(f"Chi-Square Statistic: {chi2}")
print(f"P-value: {p}")
```

```python
import pandas as pd
from sklearn.feature_selection import SelectKBest, mutual_info_classif, f_classif
data={
    'Feature1':[1,2,3,4,5],
    'Feature2': ['A','B','C','A','B'],
    'Feature3':[0,1,1,0,1],
    'Target' :[0,1,1,0,1]
}

df=pd.DataFrame(data)
X=df[['Feature1','Feature3']]
y=df['Target']

selector=SelectKBest(score_func=mutual_info_classif, k=1)
X_new = selector.fit_transform (X,y)
selected_feature_indices = selector.get_support(indices=True)
selected_features = X.columns[selected_feature_indices]
print("Selected Features:")
print(selected_features)
```

# **FEATURE SCALING**

```python
import pandas as pd
import numpy as np
import scipy as stats
from sklearn.preprocessing import StandardScaler, MinMaxScaler, Normalizer, MaxAbsScaler, RobustScaler

# Load the BMI dataset
df = pd.read_csv('/content/bmi.csv')
display(df.head())

# Clean data
df = df.dropna()

# Calculate max absolute values
max_vals = np.max(np.abs(df[['Height','Weight']]))
print(f"Max Absolute Values:\n{max_vals}")
```

### Demonstrating Different Scaling Methods

```python
# Standard Scaler
sc = StandardScaler()
df_std = df.copy()
df_std[['Height','Weight']] = sc.fit_transform(df_std[['Height','Weight']])
print("Standard Scaler Result:")
display(df_std.head())

# Min-Max Scaler
mm_scaler = MinMaxScaler()
df_mm = df.copy()
df_mm[['Height','Weight']] = mm_scaler.fit_transform(df_mm[['Height','Weight']])
print("\nMin-Max Scaler Result:")
display(df_mm.head())

# Normalizer
norm_scaler = Normalizer()
df_norm = df.copy()
df_norm[['Height','Weight']] = norm_scaler.fit_transform(df_norm[['Height','Weight']])
print("\nNormalizer Result:")
display(df_norm.head())

# Max Absolute Scaler
ma_scaler = MaxAbsScaler()
df_ma = df.copy()
df_ma[['Height','Weight']] = ma_scaler.fit_transform(df_ma[['Height','Weight']])
print("\nMax Absolute Scaler Result:")
display(df_ma.head())

# Robust Scaler
rb_scaler = RobustScaler()
df_rb = df.copy()
df_rb[['Height','Weight']] = rb_scaler.fit_transform(df_rb[['Height','Weight']])
print("\nRobust Scaler Result:")
display(df_rb.head())
```

# **KNN CLASSIFICATION (Income Dataset)**

```python
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import accuracy_score, confusion_matrix

# Load and check the dataset
data = pd.read_csv('/content/income(1) (1).csv', na_values=[" ?"])
print("Missing values summary:")
print(data.isnull().sum())

# Drop missing values
data2 = data.dropna(axis=0)

# Map target variable (handling potential case sensitivity and mapping strings to integers)
data2['SalStat'] = data2['SalStat'].map({' less than or equal to 50,000': 0, ' greater than 50,000': 1})

# Convert categorical variables to dummy variables
new_data = pd.get_dummies(data2, drop_first=True)

# Prepare Features (X) and Target (y)
columns_list = list(new_data.columns)
features = list(set(columns_list) - set(['SalStat']))

X = new_data[features].values
y = new_data['SalStat'].values

# Split into training and testing sets
train_x, test_x, train_y, test_y = train_test_split(X, y, test_size=0.3, random_state=0)

# Train KNN Classifier
KNN_classifier = KNeighborsClassifier(n_neighbors=5)
KNN_classifier.fit(train_x, train_y)

# Predictions and Evaluation
prediction = KNN_classifier.predict(test_x)
cm = confusion_matrix(test_y, prediction)
acc = accuracy_score(test_y, prediction)

print("\nConfusion Matrix:")
print(cm)
print(f"\nAccuracy Score: {acc:.4f}")
print(f"Misclassified samples: {(test_y != prediction).sum()}")
```

```python
from scipy.stats import chi2_contingency

# Chi-Square Test using Tips dataset
tips = sns.load_dataset('tips')
contingency_table = pd.crosstab(tips['sex'], tips['time'])
print("Contingency Table:")
print(contingency_table)

chi2, p, _, _ = chi2_contingency(contingency_table)
print(f"\nChi-Square Statistic: {chi2:.4f}")
print(f"P-value: {p:.4f}")
```

```python
from sklearn.feature_selection import SelectKBest, mutual_info_classif

# SelectKBest with Mutual Information on synthetic data
synthetic_data = {
    'Feature1': [1, 2, 3, 4, 5],
    'Feature2': ['A', 'B', 'C', 'A', 'B'],
    'Feature3': [0, 1, 1, 0, 1],
    'Target': [0, 1, 1, 0, 1]
}

df_syn = pd.DataFrame(synthetic_data)
X_syn = df_syn[['Feature1', 'Feature3']]
y_syn = df_syn['Target']

selector = SelectKBest(score_func=mutual_info_classif, k=1)
X_new = selector.fit_transform(X_syn, y_syn)
selected_indices = selector.get_support(indices=True)

print("Selected Features:")
print(X_syn.columns[selected_indices])
```
